```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading dataset

iris= pd.read_csv('iris.csv')
```

```python
# Printing the shape, column names, and the first few rows

print(f'Shape: {iris.shape}') # Rows x Columns
print(f'Column names: {iris.columns.tolist()}') # List of column headers
print('Head: ')
print(iris.head()) # Print first 5 rows to preview the dataset
```

```python
# Printing summary statistics

print('Info:')
print(iris.info()) # Overview of columns, data types, non-null values

print('Description:')
print(iris.describe()) # Statistical summary of numeric columns
```

# VISUALIZATION SECTION

```python
# Scatter plot  Compare Petal Length and Width, color by Species

plt.figure(figsize=(6, 4)) #6 inches(w) x 4 inches(h)
sns.scatterplot(
    x='PetalLengthCm', # X-axis feature
    y='PetalWidthCm', # Y-axis feature
    hue='Species', # Group/Color by Species
    data=iris      # Data source
)
plt.title("Petal Length vs Petal Width by Species")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.grid(True)
plt.show()
```

```python
# histrograms, showing value distributions

iris.hist(edgecolor='black', figsize=(10, 6)) # Size of entire plot grid, color of borders around bars
plt.suptitle('Feature Distributions')
plt.tight_layout()
plt.show()
```

```python
# box plots, identifying outliers

plt.figure(figsize= (10, 6))
sns.boxplot(data= iris)
plt.title('Box Plot of Features')
plt.xticks(rotation= 45)  # Tilt labels for readability
plt.show()
```
